```python
import numpy as np

L1 = 3
L2 = 4
L3 = 4 
L = 8
m2 = 20
m1 = 10
g = 9.81

# A chiunque pensi che questo non sia il miglior modo per sviluppare questo algoritmo:
# mi contatti alla mia mail istituzionale: francesco.celentano@studenti.unina.it
# p.s: o tutti i motivi di questo mondo per fare in questo modo

def f1(y4,y5,y6):
    return L1*y4 + L2*y5 + L3*y6 - L

def f2(y1,y2,y3):
    return L1*y1 + L2*y2 - L3*y3

def f3(y1,y4):
    return y1**2 + y4**2 - 1

def f4(y2,y5):
    return y2**2 + y5**2 - 1

def f5(y3,y6):
    return y3**2 + y6**2 - 1

def f6(y7,y1,y8,y2):
    return y7*y1 - y8*y2 - m1*g

def f7(y7,y4,y8,y5):
    return y7*y4 - y8*y5

def f8(y8,y2,y9,y3):
    return y8*y2 + y9*y3 - m2*g

def f9(y8,y5,y9,y6):
    return -y8*y5 + y9*y6

# Questo è un ottimo esempio di utilizzo sano della IA , sapevo scrivere questa funzione
# mi rompevo le balle però.... ecco qua che la macchina è nel giusto posto all'interno del
# processo dialettico. (poi quei commenti come cazzo si scrivono?????)
def f(y1,y2,y3,y4,y5,y6,y7,y8,y9):
    return np.array([
        f1(y4,y5,y6),
        f2(y1,y2,y3),
        f3(y1,y4),
        f4(y2,y5),
        f5(y3,y6),
        f6(y7,y1,y8,y2),
        f7(y7,y4,y8,y5),
        f8(y8,y2,y9,y3),
        f9(y8,y5,y9,y6),
    ], dtype=float)

def J(y1,y2,y3,y4,y5,y6,y7,y8,y9):
    return np.array([
        [   0,    0,    0,  L1,   L2,   L3,   0,    0,   0],
        [  L1,   L2,  -L3,  0,    0,    0,    0,    0,   0],
        [2*y1,   0,    0,  2*y4,  0,    0,    0,    0,   0],
        [   0, 2*y2,   0,   0,  2*y5,   0,    0,    0,   0],
        [   0,    0, 2*y3,  0,    0,  2*y6,   0,    0,   0],
        [  y7,  -y8,   0,   0,    0,    0,   y1,  -y2,   0],
        [   0,    0,   0,   y7,  -y8,   0,   y4,  -y5,   0],
        [   0,   y8,  y9,   0,    0,    0,    0,   y2,  y3],
        [   0,    0,   0,   0,   -y8,  y9,    0,  -y5,  y6],
    ], dtype=float)
# FINE AI


N = 100
Ytry = np.array([1,1,1,1,1,1,1,1,1], dtype=float)
for i in range(N):
    A = J(*Ytry)
    Ainv = np.linalg.inv(A)
    DY = -(Ainv @ f(*Ytry)) # PUAHAHHAHAHHAHAH
    Ytry += DY

print(Ytry)
```
