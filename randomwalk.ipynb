```python
import numpy as np
import random as ran
import matplotlib.pyplot as plt

STARTX = 50
STARTY = 50

def random_walk(MATRIX):
    i = STARTX
    j = STARTY
    passi = 0

    while(True):
        flag = ran.random()

        if(i == 0 or j == 0 or i == 99 or j == 99):
            return passi
        
        if(flag <= 0.333333):
            passi = passi + 1
            i = i - 1
            MATRIX[i][j] = 1
        elif(flag >= 0.333333 and flag <= 0.666666):
            passi = passi + 1
            j = j + 1
            MATRIX[i][j] = 1
        else:
            passi = passi + 1
            i = i + 1
            MATRIX[i][j] = 1
        
ITERAZIONI = 100

passi = np.zeros(ITERAZIONI)
CASERTA = np.zeros((100,100,ITERAZIONI))
CASERTA[STARTX][STARTY][:] = 1
for i in range(ITERAZIONI):
    passi[i] = random_walk(CASERTA[:,:,i])

print("Media: ",np.average(passi))
print("Min: ",np.min(passi))
print("Max: ",np.max(passi))
```
