```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
N=5000
nt = 50
STEP = 500
m = 1
k = 1

DT = 3*math.pi/nt
Y = np.array([2.0, 0.0])  # posizione=0, velocità=2


def g(y, t,c):
    v = np.zeros(2)
    v[0] = y[1]
    v[1] = -c/m*y[1]-k/m*y[0]
    return v


def rungeKutta(Y, t, dt,c):
    C1 = g(Y, t,c)
    C2 = g(Y + dt*C1/2, t + dt/2,c)
    C3 = g(Y + dt*C2/2, t + dt/2,c)
    C4 = g(Y + dt*C3,   t + dt,c)

    return Y + dt*(C1 + 2*C2 + 2*C3 + C4)/6
```

```python
c=1

Y1 = np.zeros(N+1)
Y2 = np.zeros(N+1)
Y1[0] = Y[0]
Y2[0] = Y[1]

for i in range(N):
    t = DT * i
    Y = rungeKutta(Y, t, DT, c)
    Y1[i+1] = Y[0]
    Y2[i+1] = Y[1]

plt.scatter(Y1, Y2)
plt.scatter(Y1,0.5*Y2**2)
plt.scatter(Y1,0.5*k*Y1)
plt.xlabel("Y1")
plt.ylabel("Y2")
plt.grid(True)
plt.show()
```
