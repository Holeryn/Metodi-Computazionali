```python
import numpy as np
import math
import matplotlib.pylab as plt
```

```python
g = 9.81
R = 0.008
m = 0.005
C = 0.5
Area = math.pi*R*R

N = 1000
dt = 0.01
rho = 1
a = 0
b = 100
```

```python
X = np.linspace(a,b,N)
V = np.zeros(N)
A = np.zeros(N)

def f(i):
    return g - C*Area*rho*V[i]**2/(2*m)

X[0] = -10
V[0] = 0
A[0] = f(0)

V[1] = V[0]+dt*(f(0))
A[1] = f(1)

for i in range(N-2):
    V[i + 2] = V[i] + 2*dt*(f(i + 1))
    V[i + 2] = V[i] + dt/3*(f(i+2) + 4*f(i+1)+f(i))
    A[i + 2] = f(i+2)

plt.scatter(X*dt,V)
plt.show()
plt.scatter(X*dt,A)
plt.show()
```
