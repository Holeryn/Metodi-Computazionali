```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 0
b = 1

N = 100
dt = 1/N
```

```python
X = np.linspace(a,b,N)
Y = np.zeros(N)
Y1 = np.zeros(N)

errori3 = np.zeros(N)
errori5 = np.zeros(N)

Y[0] = 1
Y[1] = Y[0] - dt*(X[0]*Y[0])
errori5[0] = errori3[0] = 0

Y1[0] = 1
Y1[1] = math.exp(-1/2)
errori3[1] =  Y[1] - Y1[1]
errori5[1] =  Y[1] - Y1[1]

for i in range(N-2):
    Y1[i + 2] = math.exp(-((X[i+2])**2)/2)

    # Predictor
    Y[i + 2] = Y[i] - 2*dt*X[i+1]*Y[i+1]
    errori3[i + 2] =  Y[i+2] - Y1[i+2]
    #Corrector
    Y[i + 2] = Y[i] - dt/3*(X[i+2]*Y[i+2] + 4*X[i+1]*Y[i+1] + X[i]*Y[i])
    errori5[i + 2] =  Y[i+2] - Y1[i+2]


plt.scatter(X,Y)
plt.scatter(X,Y1)
plt.show()

plt.scatter(X,errori3)
plt.show()
plt.scatter(X,errori5)
plt.show()
```
