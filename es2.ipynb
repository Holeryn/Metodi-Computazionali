```python
import matplotlib.pyplot as plt
import numpy as np
import random as ran
import math
```

```python
def f0(x):
    return x**2

def f1(x):
    return np.cos(x)   
```

```python
K = 1000                             # Numero di volte in cui si ripete l'eseperimento

def montecarlo(f,a,b,N):
    X = np.linspace(a,b,(b-a)*1000)
    F = f(X)
    FMIN = min(F)
    FMAX = max(F)

    K0 = FMIN*(b-a)
    K1 = (FMAX-FMIN)*(b-a)

    NT = N
    NS = 0
    for i in range(NT):
        x = ran.random()
        y = ran.random()

        GX = (f(a + (b-a)*x) - FMIN)/(FMAX - FMIN)
        if y <= GX:
            NS = NS + 1
    
    I0 = NS/NT
    try:
        DI0 = math.sqrt(1/NS-1/NT)
    except ZeroDivisionError:
        DI0 = 0
     
    return K1*I0+K0,K1*DI0

# Primo integrale
I10_0 , DI10_0 = np.zeros(K) , np.zeros(K)
I50_0 , DI50_0 = np.zeros(K) , np.zeros(K)
I250_0 , DI250_0 = np.zeros(K) , np.zeros(K)
I1250_0 , DI1250_0 = np.zeros(K) , np.zeros(K)
I6250_0 , DI6250_0 = np.zeros(K) , np.zeros(K)

for i in range(K):
    I10_0[i],DI10_0[i] = montecarlo(f0,0,1,10)    
    I50_0[i],DI50_0[i] = montecarlo(f0,0,1,50)    
    I250_0[i],DI250_0[i] = montecarlo(f0,0,1,250) 
    I1250_0[i],DI1250_0[i] = montecarlo(f0,0,1,1250)  
    I6250_0[i],DI6250_0[i] = montecarlo(f0,0,1,6250)  


# SECONDO integrale
I10_1 , DI10_1 = np.zeros(K) , np.zeros(K)
I50_1 , DI50_1 = np.zeros(K) , np.zeros(K)
I250_1 , DI250_1 = np.zeros(K) , np.zeros(K)
I1250_1 , DI1250_1 = np.zeros(K) , np.zeros(K)
I6250_1 , DI6250_1 = np.zeros(K) , np.zeros(K)

for i in range(K):
    I10_1[i],DI10_1[i] = montecarlo(f0,0,2,10)    
    I50_1[i],DI50_1[i] = montecarlo(f0,0,2,50)    
    I250_1[i],DI250_1[i] = montecarlo(f0,0,2,250) 
    I1250_1[i],DI1250_1[i] = montecarlo(f0,0,2,1250)  
    I6250_1[i],DI6250_1[i] = montecarlo(f0,0,2,6250)  


# TERZO integral
I10_2 , DI10_2 = np.zeros(K) , np.zeros(K)
I50_2 , DI50_2 = np.zeros(K) , np.zeros(K)
I250_2 , DI250_2 = np.zeros(K) , np.zeros(K)
I1250_2 , DI1250_2 = np.zeros(K) , np.zeros(K)
I6250_2 , DI6250_2 = np.zeros(K) , np.zeros(K)

for i in range(K):
    I10_2[i],DI10_2[i] = montecarlo(f1,0,2,10)    
    I50_2[i],DI50_2[i] = montecarlo(f1,0,2,50)    
    I250_2[i],DI250_2[i] = montecarlo(f1,0,2,250) 
    I1250_2[i],DI1250_2[i] = montecarlo(f1,0,2,1250)  
    I6250_2[i],DI6250_2[i] = montecarlo(f1,0,2,6250)  

print("primo ingegrale: ",np.average(I6250_0))
print("Secondo ingegrale: ",np.average(I6250_1))
print("terzo ingegrale: ",np.average(I6250_2))
```

```python
plt.title("primo")
plt.scatter([np.std(I10_0),np.std(I50_0),np.std(I250_0),np.std(I1250_0),np.std(I6250_0)],[1/math.sqrt(10),1/math.sqrt(50),
                                                       1/math.sqrt(250),1/math.sqrt(1250),1/math.sqrt(6250)])
plt.show()

plt.title("SECONDO")
plt.scatter([np.std(I10_1),np.std(I50_1),np.std(I250_1),np.std(I1250_1),np.std(I6250_1)],[1/math.sqrt(10),1/math.sqrt(50),
                                                       1/math.sqrt(250),1/math.sqrt(1250),1/math.sqrt(6250)])
plt.show()

plt.title("TERZO")
plt.scatter([np.std(I10_2),np.std(I50_2),np.std(I250_2),np.std(I1250_2),np.std(I6250_2)],[1/math.sqrt(10),1/math.sqrt(50),
                                                       1/math.sqrt(250),1/math.sqrt(1250),1/math.sqrt(6250)])
plt.show()

plt.title("sovrapposti")
plt.scatter([np.std(I10_0),np.std(I50_0),np.std(I250_0),np.std(I1250_0),np.std(I6250_0)],[1/math.sqrt(10),1/math.sqrt(50),
                                                       1/math.sqrt(250),1/math.sqrt(1250),1/math.sqrt(6250)])
plt.scatter([np.std(I10_1),np.std(I50_1),np.std(I250_1),np.std(I1250_1),np.std(I6250_1)],[1/math.sqrt(10),1/math.sqrt(50),
                                                       1/math.sqrt(250),1/math.sqrt(1250),1/math.sqrt(6250)])
plt.scatter([np.std(I10_2),np.std(I50_2),np.std(I250_2),np.std(I1250_2),np.std(I6250_2)],[1/math.sqrt(10),1/math.sqrt(50),
                                                       1/math.sqrt(250),1/math.sqrt(1250),1/math.sqrt(6250)])
plt.show()
```
