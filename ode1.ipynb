```python
import math
import numpy as np
import matplotlib.pyplot as plt

a = 0
b = 1

N = 100
dt = 1/N

X = np.linspace(a,b,N)
Y = np.zeros(N)
Y1 = np.zeros(N)
Y[0] = 1
Y1[0] = 1
for i in range(N - 1):
    Y[i + 1] = Y[i] + dt*(X[i] + Y[i])
    Y1[i+1] = -(X[i] + 1) + 2*math.exp(X[i])

plt.scatter(X,Y)
plt.scatter(X,Y1)

error = np.zeros(N)
error = np.abs(Y-Y1)
erroravg = np.average(error)
print(erroravg)
```
